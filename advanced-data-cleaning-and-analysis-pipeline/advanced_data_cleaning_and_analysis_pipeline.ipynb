## Task 1: Create Reusable Data Cleaning Function

Abstract previous cleaning logic into a parameterized function:
- Create `clean_user(user_info, name_index, age_index)` function
- Strip whitespace and replace underscores in names
- Convert age from float to integer
- Split full names into list components
- Return modified user record without mutating original data
- Test with sample user record


````python
# Define your function here
def clean_user(user_info, name_index, age_index):
    # Step 1: Remove leading/trailing spaces and underscores from name
    user_name_1 = user_info[name_index].strip().replace('_', ' ')

    # Step 2: Convert age to integer
    user_age_1 = int(user_info[age_index])

    # Step 3: Split name and last name into a sublist
    user_name_1 = user_name_1.split() 

    # Prepare list with complete user information
    # Replace original name and age with cleaned data
    user_info[name_index] = user_name_1
    user_info[age_index] = user_age_1

    return user_info  # Return with corrected values

# Test the function
test_user = ['32415', ' mike_reed ', 32.0, ['ELECTRONICS', 'SPORT', 'BOOKS'], [894, 213, 173]]
name_index = 1
age_index = 2

print(clean_user(test_user, name_index, age_index))
````

**Implementation Notes:**
- Use `strip()` and `replace()` for string normalization
- Convert age using `int()` function
- Use `split()` to create name sublist
- Return deep copy to preserve original data

## Task 10: Generic Category-Based Reporting Function

Build reusable function for flexible customer queries:

**Function Signature:** `get_client_by_cat(users, id_index, name_index, age_index, category_index, amounts_index, filter_category)`

**Parameters:**
- `users`: Customer dataset (nested lists)
- `id_index`: Position of customer ID
- `name_index`: Position of name components
- `age_index`: Position of age value
- `category_index`: Position of category list
- `amounts_index`: Position of spending amounts
- `filter_category`: Target category for filtering

**Output Format:**
```python
[[customer_id, [first_name, last_name], age, total_spent], ...]
```

**Example:**
```python
get_client_by_cat(users, 0, 1, 2, 3, 4, 'books')
# Returns: [['32415', ['mike', 'reed'], 32, 1280]]
```

Apply function to provided dataset filtered by 'home' category.


````python
def get_client_by_cat(users, id_index, name_index, age_index, category_index, amounts_index, filter_category):
    # Write your code here
    
    result = []
    
    for user in users:
        if filter_category in user[category_index]:  # If filter_category is in user category list
            total_spent = sum(user[amounts_index])  # Calculate total spending from spending list
            result.append([user[id_index], user[name_index], user[age_index], total_spent])  # Append data to results list
            # Results contain: (id, name, age, total_spent, filter_category)
            
    return result
            
# User list
users = [
    ['32415', ['mike', 'reed'], 32, ['electronics', 'sport', 'books'], [894, 213, 173]],
    ['31980', ['kate', 'morgan'], 24, ['clothes', 'books'], [439, 390]],
    ['32156', ['john', 'doe'], 37, ['electronics', 'home', 'food'], [459, 120, 99]],
    ['32761', ['samantha', 'smith'], 29, ['clothes', 'electronics', 'beauty'], [299, 679, 85]],
    ['32984', ['david', 'white'], 41, ['books', 'home', 'sport'], [234, 329, 243]],
    ['33001', ['emily', 'brown'], 26, ['beauty', 'home', 'food'], [213, 659, 79]],
    ['33767', ['maria', 'garcia'], 33, ['clothes', 'food', 'beauty'], [499, 189, 63]],
    ['33912', ['jose', 'martinez'], 22, ['sport', 'electronics', 'home'], [259, 549, 109]],
    ['34009', ['lisa', 'wilson'], 35, ['home', 'books', 'clothes'], [329, 189, 329]],
    ['34278', ['james', 'lee'], 28, ['beauty', 'clothes', 'electronics'], [189, 299, 579]]
]

# Call function with 'home' category
result = get_client_by_cat(users, 0, 1, 2, 3, 4, 'home')  # Indices: (0=id, 1=name, 2=age, 3=categories, 4=amounts)

# Display result list
print(result)
````
